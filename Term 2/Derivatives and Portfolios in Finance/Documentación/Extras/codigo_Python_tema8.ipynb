Sea el siguiente problema

\begin{equation*}
\left\lbrace
\begin{array}{ccl}
	\text{mín}  &      & R(x)=x^{T} Ax\\
	\text{s.a.} &     &\sum\limits_{i=1}^n{x_i}=M,\\
	&     &\sum\limits_{i=1}^n{\mu_ix_i}=\mu,\\
	&     & x_i\geq 0,\,\forall i\in\{1,\ldots,n\}.
\end{array}
\right.
\end{equation*}

Cuya forma matricial es
\begin{equation*}
\left\lbrace
\begin{array}{ccl}
	\min &      & x^{T} A x\\
	\text{s.a.} &     & Bx = b,\\
	&     & x\geq0,\\
\end{array}
\right.
\end{equation*}

Entonces

\begin{equation*}
\lambda^{\ast}= 2(B A^{-1} B^{T})^{-1}b,
\end{equation*}
Por lo tanto
\begin{equation*}
x^{\ast}
=\frac12
A^{-1}B^{T} \lambda^{\ast}.
\end{equation*}


# Ejemplo

```python
import numpy as np

# Definición de matrices y vectores
A = np.array([[2, 0, 0],
              [0, 1, 0], 
              [0, 0, 3]])
B = np.array([[1, 1, 1], 
              [5, 10, 15]])
b = np.array([[14],
              [150]])

# Inversa de A para evitar múltiples cálculos
iA = np.linalg.inv(A)

# Uso de operaciones matriciales para calcular last
# `@` es el operador de multiplicación matricial en Python
# np.linalg.inv calcula la inversa de la matriz
# Evitamos calcular la inversa de B@iA@B^T más de una vez almacenándola en una variable
BAB_inv = np.linalg.inv(B @ iA @ B.T)
l_ast = 2 * BAB_inv @ b

# Cálculo de x_ast usando la inversa precalculada de A y los resultados de 'last'
x_ast = iA @ B.T @ l_ast / 2

# Impresión de resultados con redondeo a 5 decimales
print('lambda_op:', np.round(l_ast.T, 5))
print('x_op:', np.round(x_ast.T, 5))
```
