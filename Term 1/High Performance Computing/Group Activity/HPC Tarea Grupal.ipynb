## Tarea 1. Interpolación polinómica

### Explicación del código:

Este código realiza la interpolación de tres funciones utilizando diferentes métodos: interpolación baricéntrica, interpolación de Lagrange y diferencias divididas de Newton. Aquí está la explicación de las partes más importantes del código:

#### Importación de bibliotecas:

Se importa la biblioteca NumPy como np para realizar operaciones numéricas y manipulación de matrices.
Se importa el módulo scipy.interpolate de SciPy para realizar la interpolación.
Se importa el módulo scipy.linalg de SciPy para realizar cálculos de álgebra lineal.
Se importa la biblioteca time para medir el tiempo de ejecución.
Se importa la biblioteca matplotlib.pyplot como plt para generar gráficos.

#### Definición de las funciones a interpolar:

Se definen tres funciones f1, f2 y f3 que representan las funciones a interpolar. Estas funciones están definidas utilizando expresiones lambda.

#### Definición de parámetros:

num_nodos es una lista que contiene el número de nodos para realizar la interpolación.
a y b representan los límites izquierdo y derecho del intervalo en el que se realizará la interpolación.

#### Funciones auxiliares:

obtener_nodos_equiespaciados es una función que devuelve un arreglo de nodos equiespaciados dentro del intervalo [a, b].
obtener_nodos_chebyshev es una función que devuelve un arreglo de nodos de Chebyshev escalados dentro del intervalo [a, b].

#### Métodos de interpolación:

interpolacion_baricentrica es una función que realiza la interpolación baricéntrica utilizando los nodos y puntos de      evaluación dados.
interpolacion_lagrange es una función que realiza la interpolación de Lagrange utilizando los nodos y puntos de evaluación dados.
diferencias_divididas es una función que calcula la tabla de diferencias divididas de Newton para los nodos y puntos de la función dados.
polinomio_newton es una función que evalúa el polinomio de Newton utilizando los coeficientes de las diferencias divididas, los nodos y un valor x.

#### Interpolación y visualización:

Se define un rango de valores x_range para las gráficas.
El código realiza la interpolación para cada número de nodos en num_nodos.
Para cada número de nodos, se obtienen los nodos equiespaciados y los nodos de Chebyshev utilizando las funciones auxiliares.
Se itera sobre las funciones y los nodos para realizar la interpolación y calcular el error de interpolación.
Se generan gráficas para mostrar los resultados de la interpolación y se muestran en pantalla.

En resumen, este código muestra cómo realizar la interpolación de funciones utilizando diferentes métodos y compara los resultados. También calcula el error de interpolación y muestra gráficas para visualizar los resultados.

```python
import numpy as np
import scipy.interpolate as interp
import scipy.linalg as linalg
import time
import matplotlib.pyplot as plt

# Funciones a interpolar
f1 = lambda x: np.sin(x)
f2 = lambda x: 1 / (1 + 25 * x**2)
f3 = lambda x: np.exp(-20 * x**2)

# Número de nodos
num_nodos = [11, 21]
# Limite izquierdo del intervalo
a = -np.pi/2
# Limite derecho del intervalo
b = np.pi/2

# Obtención de los nodos equiespaciados
def obtener_nodos_equiespaciados(a, b, num_nodos):
    return np.linspace(a, b, num_nodos)

# Obtención de los nodos de Chebyshev
def obtener_nodos_chebyshev(a, b, num_nodos):
    nodos_chebyshev = np.polynomial.chebyshev.Chebyshev.basis(num_nodos).roots()
    nodos_escalados = 0.5 * (b - a) * nodos_chebyshev + 0.5 * (b + a)
    return nodos_escalados

# Método de interpolación baricéntrica
def interpolacion_baricentrica(x, y, puntos_evaluacion):
    interpolacion = interp.barycentric_interpolate(x, y, puntos_evaluacion)
    return interpolacion
# Método de interpolación de Lagrange
def interpolacion_lagrange(x, y, puntos_evaluacion):
    polinomio = interp.lagrange(x, y)
    interpolacion = polinomio(puntos_evaluacion)
    return interpolacion

# Método de diferencias divididas de Newton
def diferencias_divididas(x, y): #    funcion para calcular la tabla de diferencias finitas
    n = len(y)
    coef = np.zeros([n, n])
    # the first column is y
    coef[:,0] = y
    
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = \
           (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
            
    return coef

def polinomio_newton(coef, x_data, x): # Funcion que evalua el polinomio de newton
    n = len(x_data) - 1 
    p = coef[n]
    for k in range(1,n+1):
        p = coef[n-k] + (x -x_data[n-k])*p
    return p

# Rango de valores x para las gráficas
x_range = np.linspace(a, b, 500)

# Interpolación para cada función, método y número de nodos
for num_puntos in num_nodos:
    print(f"--- Número de nodos: {num_puntos} ---")

    # Obtención de los nodos equiespaciados
    nodos_equiespaciados = obtener_nodos_equiespaciados(a, b, num_puntos)
    
    # Obtención de los nodos de Chebyshev
    nodos_chebyshev = obtener_nodos_chebyshev(a, b, num_puntos)
    
    for i, (nodos, nodos_label) in enumerate([(nodos_equiespaciados, "Equiespaciados"), (nodos_chebyshev, "Chebyshev")]):
        print(f"\nMétodo de nodos {nodos_label}:")
        print(f"\nNodos x: {nodos}")

        # Puntos de la función para los nodos de interpolación
        puntos_funcion = {
            "f1": f1(nodos),
            "f2": f2(nodos),
            "f3": f3(nodos)
        }

        for j, (funcion_label, funcion) in enumerate(puntos_funcion.items()):
            print(f"\nFunción: {funcion_label}")
            print("Puntos de la función:", funcion)

            # Interpolación mediante método de interpolación baricéntrica
            start_time = time.time()
            interpolacion_bar = interpolacion_baricentrica(nodos, funcion, x_range)
            elapsed_time = time.time() - start_time
            print(f"\nMétodo baricéntrico: {elapsed_time:.6f} segundos")

            # Interpolación de la función en el rango completo
            funcion_interpolada = interp.barycentric_interpolate(nodos, funcion, x_range)

            # Cálculo del error de interpolación
            error_bar = np.linalg.norm(funcion_interpolada - f1(x_range))
            print(f"Error de interpolación (baricéntrico): {error_bar:.6f}")
            
            if j==0:
                f_range = f1(x_range)
            if j==1: 
                f_range = f2(x_range)
            if j==2:
                f_range = f3(x_range)
            

            # Gráfica de la función y la interpolación
            plt.figure()
            plt.plot(nodos, funcion, 'or',label="Puntos")
            plt.plot(x_range, f_range, label="Función Real")
            plt.plot(x_range, funcion_interpolada, label="Interpolación Baricéntrica")
            plt.xlabel("x")
            plt.ylabel("y")
            plt.title(f"Interpolación Baricentrica para {funcion_label} ({num_puntos} nodos - {nodos_label})")
            plt.legend()
            plt.grid()
            plt.show()
            
            # Interpolación mediante método de interpolación de Lagrange
            start_time = time.time()
            interpolacion_lag = interpolacion_lagrange(nodos, funcion, x_range)
            elapsed_time = time.time() - start_time
            print(f"\nMétodo de Lagrange: {elapsed_time:.6f} segundos")
            error_lag = np.linalg.norm(interpolacion_lag - f_range)
            print(f"Error de interpolación (Lagrange): {error_lag:.6f}")
            
            # Gráfica de la función y la interpolación de Lagrange
            plt.figure()
            plt.plot(nodos, funcion, 'or',label="Puntos")
            plt.plot(x_range, f_range, label="Función Real")
            plt.plot(x_range, interpolacion_lag, label="Interpolación Lagrange")
            plt.xlabel("x")
            plt.ylabel("y")
            plt.title(f"Interpolación de Lagrange para {funcion_label} ({num_puntos} nodos - {nodos_label})")
            plt.legend()
            plt.grid()
            plt.show()
            
             # Interpolación mediante método de diferencias divididas de Newton
            start_time = time.time()
            a_s = diferencias_divididas(nodos, funcion)[0, :]
            y_new = polinomio_newton(a_s, nodos, x_range)
            elapsed_time = time.time() - start_time
            print(f"\nMétodo de diferencias divididas de Newton: {elapsed_time:.6f} segundos")
            error_newt = np.linalg.norm(y_new - f_range)
            print(f"Error de interpolación (Newton): {error_newt:.6f}")
            
            # Gráfica de la función y la interpolacion de Newton por diferencias divididas
            plt.figure()
            plt.plot(nodos, funcion, 'or',label="Puntos")
            plt.plot(x_range, f_range, label="Función Real")
            plt.plot(x_range, y_new, label="Interpolación Newton")
            plt.xlabel("x")
            plt.ylabel("y")
            plt.title(f"Interpolación por diferencias divididas para {funcion_label} ({num_puntos} nodos - {nodos_label})")
            plt.legend()
            plt.grid()
            plt.show()
            
            
```

## Tarea 2. Interpolación con splines

### Explicación del código:

#### Importación de bibliotecas:

Se importan las bibliotecas necesarias: numpy, matplotlib.pyplot, scipy.interpolate.InterpolatedUnivariateSpline y numpy.polynomial.chebyshev.

#### Definición de las funciones a interpolar:

Se definen tres funciones: f1, f2 y f3, que representan las funciones a interpolar.

#### Definición de parámetros:

Se establece el número de nodos de interpolación deseado (11 en este caso) y se generan los nodos de interpolación utilizando el método de Chebyshev.

#### interpolación con splines

Se evalúan las funciones en los nodos de interpolación para obtener los puntos correspondientes.
Se realiza la interpolación de cada función utilizando splines cúbicos, creando los objetos spline_11_f1, spline_11_f2 y spline_11_f3.
Se calcula el error de interpolación para cada función, comparando los valores interpolados con los valores reales en un conjunto de puntos x_vals.
Se calcula el tiempo de cálculo de los polinomios interpoladores para cada función.
Se generan las gráficas que muestran la función real y el polinomio obtenido mediante la interpolación para cada función.
Se imprime el error de interpolación y el tiempo de interpolación para cada función y número de nodos (11 y 21).
Se repiten los pasos del 3 al 9 utilizando 21 nodos de interpolación en lugar de 11.
Se obtienen los polinomios interpoladores para la función con 11 nodos de interpolación.
Se imprimen los polinomios interpoladores.

En resumen, el código realiza la interpolación de tres funciones utilizando splines cúbicos y proporciona información sobre el error de interpolación, el tiempo de cálculo y los polinomios interpoladores. Además, genera gráficas para visualizar los resultados de la interpolación.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import InterpolatedUnivariateSpline
from numpy.polynomial import chebyshev

# Definición de las funciones
def f1(x):
    return np.sin(x)

def f2(x):
    return 1 / (1 + 25 * x**2)

def f3(x):
    return np.exp(-20 * x**2)

# Obtención de los nodos de interpolación usando Chebyshev
num_nodes = 11  # Número de nodos de interpolación
x_nodes_11 = chebyshev.chebpts1(num_nodes)  # Nodos de interpolación

# Obtención de los puntos de la función en los nodos de interpolación
y_nodes_11_f1 = f1(x_nodes_11)
y_nodes_11_f2 = f2(x_nodes_11)
y_nodes_11_f3 = f3(x_nodes_11)

# Interpolación con splines cúbicos
spline_11_f1 = InterpolatedUnivariateSpline(x_nodes_11, y_nodes_11_f1)
spline_11_f2 = InterpolatedUnivariateSpline(x_nodes_11, y_nodes_11_f2)
spline_11_f3 = InterpolatedUnivariateSpline(x_nodes_11, y_nodes_11_f3)

# Cálculo del error de interpolación
x_vals = np.linspace(-1, 1, 1000)

error_11_f1 = np.abs(f1(x_vals) - spline_11_f1(x_vals)).max()
error_11_f2 = np.abs(f2(x_vals) - spline_11_f2(x_vals)).max()
error_11_f3 = np.abs(f3(x_vals) - spline_11_f3(x_vals)).max()

# Tiempos de cálculo de los polinomios
import time

start_time = time.time()
spline_11_f1(x_vals)
end_time = time.time()
time_11_f1 = end_time - start_time

start_time = time.time()
spline_11_f2(x_vals)
end_time = time.time()
time_11_f2 = end_time - start_time

start_time = time.time()
spline_11_f3(x_vals)
end_time = time.time()
time_11_f3 = end_time - start_time

# Gráficas de la función real y del polinomio obtenido
plt.figure(figsize=(10, 5))

# Función f(x) = sin(x)
plt.subplot(131)
plt.plot(x_vals, f1(x_vals), label='Función Real')
plt.plot(x_vals, spline_11_f1(x_vals), label='Spline 11 Nodos')
plt.scatter(x_nodes_11, y_nodes_11_f1, color='red', label='Puntos 11 Nodos')
plt.title('f(x) = sin(x)')
plt.legend()

# Función f(x) = 1 / (1 + 25x^2)
plt.subplot(132)
plt.plot(x_vals, f2(x_vals), label='Función Real')
plt.plot(x_vals, spline_11_f2(x_vals), label='Spline 11 Nodos')
plt.scatter(x_nodes_11, y_nodes_11_f2, color='red', label='Puntos 11 Nodos')
plt.title('f(x) = 1 / (1 + 25x^2)')
plt.legend()

# Función f(x) = e^(-20x^2)
plt.subplot(133)
plt.plot(x_vals, f3(x_vals), label='Función Real')
plt.plot(x_vals, spline_11_f3(x_vals), label='Spline 11 Nodos')
plt.scatter(x_nodes_11, y_nodes_11_f3, color='red', label='Puntos 11 Nodos')
plt.title('f(x) = e^(-20x^2)')
plt.legend()

plt.tight_layout()
plt.show()

# Cálculo del error de interpolación
print("Errores de Interpolación:")
print("f(x) = sin(x):")
print("- 11 Nodos:", format(error_11_f1, '.5f'))
print()
print("f(x) = 1 / (1 + 25x^2):")
print("- 11 Nodos:", format(error_11_f2, '.5f'))
print()
print("f(x) = e^(-20x^2):")
print("- 11 Nodos:", format(error_11_f3, '.5f'))

print()

# Tiempos de Interpolación
print("Tiempos de Interpolación:")
print("f(x) = sin(x):")
print("- 11 Nodos:", format(time_11_f1, '.5f'))
print()
print("f(x) = 1 / (1 + 25x^2):")
print("- 11 Nodos:", format(time_11_f2, '.5f'))
print()
print("f(x) = e^(-20x^2):")
print("- 11 Nodos:", format(time_11_f3, '.5f'))



# Obtención de los nodos de interpolación usando Chebyshev
num_nodes = 21  # Número de nodos de interpolación
x_nodes_21 = chebyshev.chebpts1(num_nodes)  # Nodos de interpolación

# Obtención de los puntos de la función en los nodos de interpolación
y_nodes_21_f1 = f1(x_nodes_21)
y_nodes_21_f2 = f2(x_nodes_21)
y_nodes_21_f3 = f3(x_nodes_21)

# Interpolación con splines cúbicos
spline_21_f1 = InterpolatedUnivariateSpline(x_nodes_21, y_nodes_21_f1)
spline_21_f2 = InterpolatedUnivariateSpline(x_nodes_21, y_nodes_21_f2)
spline_21_f3 = InterpolatedUnivariateSpline(x_nodes_21, y_nodes_21_f3)

# Cálculo del error de interpolación
x_vals = np.linspace(-1, 1, 1000)

error_21_f1 = np.abs(f1(x_vals) - spline_21_f1(x_vals)).max()
error_21_f2 = np.abs(f2(x_vals) - spline_21_f2(x_vals)).max()
error_21_f3 = np.abs(f3(x_vals) - spline_21_f3(x_vals)).max()

# Tiempos de cálculo de los polinomios
import time

start_time = time.time()
spline_21_f1(x_vals)
end_time = time.time()
time_21_f1 = end_time - start_time

start_time = time.time()
spline_21_f2(x_vals)
end_time = time.time()
time_21_f2 = end_time - start_time

start_time = time.time()
spline_21_f3(x_vals)
end_time = time.time()
time_21_f3 = end_time - start_time

# Gráficas de la función real y del polinomio obtenido
plt.figure(figsize=(10, 5))

# Función f(x) = sin(x)
plt.subplot(131)
plt.plot(x_vals, f1(x_vals), label='Función Real')
plt.plot(x_vals, spline_21_f1(x_vals), label='Spline 21 Nodos')
plt.scatter(x_nodes_21, y_nodes_21_f1, color='red', label='Puntos 21 Nodos')
plt.title('f(x) = sin(x)')
plt.legend()

# Función f(x) = 1 / (1 + 25x^2)
plt.subplot(132)
plt.plot(x_vals, f2(x_vals), label='Función Real')
plt.plot(x_vals, spline_21_f2(x_vals), label='Spline 21 Nodos')
plt.scatter(x_nodes_21, y_nodes_21_f2, color='red', label='Puntos 21 Nodos')
plt.title('f(x) = 1 / (1 + 25x^2)')
plt.legend()

# Función f(x) = e^(-20x^2)
plt.subplot(133)
plt.plot(x_vals, f3(x_vals), label='Función Real')
plt.plot(x_vals, spline_21_f3(x_vals), label='Spline 21 Nodos')
plt.scatter(x_nodes_21, y_nodes_21_f3, color='red', label='Puntos 21 Nodos')
plt.title('f(x) = e^(-20x^2)')
plt.legend()

plt.tight_layout()
plt.show()

# Cálculo del error de interpolación
print("Errores de Interpolación:")
print("f(x) = sin(x):")
print("- 21 Nodos:", format(error_21_f1, '.5f'))
print()
print("f(x) = 1 / (1 + 25x^2):")
print("- 21 Nodos:", format(error_21_f2, '.5f'))
print()
print("f(x) = e^(-20x^2):")
print("- 21 Nodos:", format(error_21_f3, '.5f'))

print()

# Tiempos de Interpolación
print("Tiempos de Interpolación:")
print("f(x) = sin(x):")
print("- 21 Nodos:", format(time_21_f1, '.5f'))
print()
print("f(x) = 1 / (1 + 25x^2):")
print("- 21 Nodos:", format(time_21_f2, '.5f'))
print()
print("f(x) = e^(-20x^2):")
print("- 21 Nodos:", format(time_21_f3, '.5f'))


# Obtención del polinomio interpolador
polynomial_11_f1 = np.poly1d(spline_11_f1.get_coeffs())
polynomial_11_f2 = np.poly1d(spline_11_f2.get_coeffs())
polynomial_11_f3 = np.poly1d(spline_11_f3.get_coeffs())

# Impresión de los polinomios interpoladores
print("Polinomio interpolador:")
print("f(x) = sin(x):")
print(polynomial_11_f1)
print()
print("f(x) = 1 / (1 + 25x^2):")
print(polynomial_11_f2)
print()
print("f(x) = e^(-20x^2):")
print(polynomial_11_f3)

```
