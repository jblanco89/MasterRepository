## Ejercicio 3

Define una función dentro de la clase Numero que a partir de una cadena de caracteres, devuelve True si esa cadena de caracteres corresponde con un número romano y falso en caso contrario. Después, modifica el método para que lance un error en el caso de que el valor que nos pasan por parámetro no se corresponde con el patrón de un número romano

```python
class Numero:
  def __init__(self, numero):
    self.normal = numero
    self.romano = self.intToRoman(numero)

  def intToRoman(self, num): 
    # taken from: https://www.geeksforgeeks.org/python-program-to-convert-integer-to-roman/
    # Storing roman values of digits from 0-9
    # when placed at different places
    m = ["", "M", "MM", "MMM"]
    c = ["", "C", "CC", "CCC", "CD", "D",
         "DC", "DCC", "DCCC", "CM "]
    x = ["", "X", "XX", "XXX", "XL", "L",
         "LX", "LXX", "LXXX", "XC"]
    i = ["", "I", "II", "III", "IV", "V",
         "VI", "VII", "VIII", "IX"]
 
    # Converting to roman
    thousands = m[num // 1000]
    hundreds = c[(num % 1000) // 100]
    tens = x[(num % 100) // 10]
    ones = i[num % 10]
 
    ans = (thousands + hundreds +
           tens + ones)
 
    return ans
  
  def romanToInt(self, s):
      #taken from:https://www.tutorialspoint.com/roman-to-integer-in-python
      """
      :type s: str
      :rtype: int
      """
      roman = {'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000,'IV':4,'IX':9,'XL':40,'XC':90,'CD':400,'CM':900}
      i = 0
      num = 0
      while i < len(s):
         if i+1<len(s) and s[i:i+2] in roman:
            num+=roman[s[i:i+2]]
            i+=2
         else:
            #print(i)
            num+=roman[s[i]]
            i+=1
      return num

  def imprime(self):
    return print(self.normal, self.romano)
    
  def suma_romano(self,roman_input:str):
    romano2 = self.romanToInt(roman_input)
    romano1 = self.normal
    suma = romano1 + romano2
    self.romano = self.intToRoman(suma)
    return suma
  
  def is_romano(self,string:str):
    import re
    check_n = bool(re.search(r"^M{0,3}(CM|CD|D?C{0,3})(XC|XL|L?X{0,3})(IX|IV|V?I{0,3})$",string))
    return check_n

    
```
